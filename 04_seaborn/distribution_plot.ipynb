Univariate

```python
import numpy as np
import pandas as py
import matplotlib.pyplot as plt
import seaborn as sb
tips = sb.load_dataset("tips")
tips
```

```python
sb.set_style("darkgrid")
```

01 histplot()

```python
sb.histplot(x= "total_bill",data=tips)
```

```python
# bins=  :set size of histogram 
sb.histplot(x= "total_bill",data=tips,bins=30)
```

02 Kdeplot

```python
sb.kdeplot(x= "tip",data=tips)
```

03 displot()

```python
sb.displot(tips["total_bill"])
```

04 countplot

```python
sb.countplot(x='total_bill',data=tips)
```

```python
sb.countplot(x='size',data=tips)
```
