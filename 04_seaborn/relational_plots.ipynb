Bivariate

```python
import numpy as np
import pandas as py
import matplotlib.pyplot as plt
import seaborn as sb
tips = sb.load_dataset("tips")
tips
```

01 jointplot

```python
sb.set_style("darkgrid")
sb.jointplot(x= "total_bill",y="tip",data=tips)
```

02 kind= " hex " , " kde " ," hist "


```python
sb.jointplot(x= "total_bill",y= "tip",data=tips,kind="hex")
```

```python
sb.jointplot(x= "total_bill",y= "tip",data=tips,kind="hist")
```

```python
sb.jointplot(x= "total_bill",y= "tip",data=tips,kind="kde")
```

03 lmplot()

```python
sb.lmplot(x='total_bill',y='tip',data=tips)
```

04 sb.pairplot

```python
# prints all the graphs
sb.pairplot(tips)
```

```python
sb.pairplot(tips,hue="sex")
```

05 boxplot()

```python
plt.figure(figsize=(10,8))
sb.boxplot(x='day', y='total_bill', data=tips)
```
