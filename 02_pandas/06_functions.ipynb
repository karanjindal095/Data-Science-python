01 Missing values

```python
dis = {"A":[1,2,3,np.nan,np.nan],
       "B":[5,np.nan,np.nan,np.nan,np.nan],
       "C":[10,20,30,40,50],
       "D":[np.nan,np.nan,np.nan,np.nan,np.nan]
       }
new = pd.DataFrame(dis)
new
```

```python
new.head(3)
```

```python
#01 isnull()
new.isnull()
```

```python
#02 isnull().sum()
new.isnull().sum()
```

```python
#03 dropna()  :gives null(R/C) column if axis is 1 and if axis is 0 then row
new1= new.dropna(axis=1)
new1
```

```python
# 04 dropna(axis = 1,thresh = 2)
new3 =new.dropna(axis=1,thresh=1)
new3
```

```python
# 05 fillna() :it will fill the Nan values
new.fillna("fill")
```

```python
new.fillna(new.mean())  #:it will fill NAn(fill) with mean,median,mode
```

```python
new.fillna(0)
```

02 Grouping

```python
d = {"Company":["FB", "GOOGLE", "MICROSOFT", "FB", "GOOGLE", "FB", "MICROSOFT", "FB"],
    "Employee":["Sam", "Rachel", "Maddy", "Joe", "Srishti", "Shivay", "Pushpa", "Kirti"],
    "Sales":[1000, 500, 550, 2000, 890, 500, 350, 350]}

df1 = pd.DataFrame(d)
df1
```

```python
df1.min()
```

```python
df1.max()
```

```python
# groupedby("column")  :replace (row's index) with column 
grouped = df1.groupby("Company")
grouped
```

```python
grouped.min()
```

```python
grouped.max()
```

```python
grouped.describe()
```

```python
df1.describe()
```

03 Custom Functions

```python
# apply()   :it applies function to the dataframe
def bonus(sales):
    return sales + 100

df1['Sales'].apply(bonus)
```

```python
df1['Sales'] = df1['Sales'].apply(lambda x: x+ 100)
df1
```

04 Joining

```python
new_e= pd.DataFrame({'Company':['GOOGLE'], 'Employee':['Kriti'], 'Sales':[5000]})
new_e
```

```python
# concate() :it concatinates the two dataframes
df1 = pd.concat([df1,new_e])
df1
```
