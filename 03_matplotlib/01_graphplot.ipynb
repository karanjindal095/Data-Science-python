02 (A) marker    = "o" , " * " , "."   

  (B) linestlye = "dotted" , " dashed " , " dashdot "

   (C) color     = "r" , "hotpink" , "g"

   (D) linewidth = "15.0"

```python
# (A) Marker
y = np.array([3,8,1,10])
plt.plot(y,marker= "o")
```

```python
y = np.array([3,8,1,10])
plt.plot(y,marker= "*")
```

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = ".")
```

```python
# (B) linestyle
y = np.array([3,8,1,10])
plt.plot(y,marker = "o",linestyle= "dotted")
```

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = "o",linestyle= "dashed")
```

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = "o",linestyle= "dashdot")
```

```python
# (C) color
y = np.array([3,8,1,10])
plt.plot(y,marker = "o",linestyle= "dashdot",color= "r")
```

```python
y = np.array([3,8,1,10])
plt.plot(y,marker = "o",linestyle= "dashdot",color= "hotpink")
```

```python
# (D) linewidth
y = np.array([3,8,1,10])
plt.plot(y,marker = "o",linestyle= "dotted",color= "r",linewidth= "10.0")
```

04 plt.subplot(row,column,index of graph) : index starts with 1

```python
plt.subplot(1,4,1)
plt.plot(x,y,"r--")
plt.subplot(1,4,2)
plt.plot(x,y,"g*-")
plt.subplot(1,4,3)
plt.plot(x,y,"y--")
plt.subplot(1,4,4)
plt.plot(x,y,"b*-")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"r*-")
plt.subplot(2,2,2)
plt.plot(x,y,"g--")
plt.subplot(2,2,3)
plt.plot(x,y,"y*-")
plt.subplot(2,2,4)
plt.plot(x,y,"--",color= "hotpink")
```

05 pie-chart

```python
sizes = [250,150,200,300,100]
labels = ["Java","c","c++","Python","others"]
explode = [0,0.4,0,0,0]
plt.pie(sizes,labels=labels,explode=explode)
```

```python
sizes = [225,150,200,250]
labels =["Dell","others","HP","Asus"]
explode = [0,0.4,0,0]
plt.pie(sizes,labels=labels,explode=explode,autopct= "%1.2f%%") 
```

06 Bar-graph

```python
x = [2,4,6]
y = [4,6,8]
x2 = [9,11,13] 
y2 = [5,7,9]
plt.bar(x,y) 
plt.bar(x2,y2)
plt.title("Bar-graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# horizontal bar-graph
x=[2,4,8]
y=[1,3,7]
x2=[11,13,15]
y2=[12,14,16]
plt.barh(x,y)
plt.barh(x2,y2)
plt.title("bar graph")
plt.xlabel('X axis')
plt.ylabel('Y axis')
```
