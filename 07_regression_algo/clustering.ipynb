```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i)

  km.fit_predict(df)
  wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```
